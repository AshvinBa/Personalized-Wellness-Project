```python
import pandas as pd
```

```python
df=pd.read_csv("Final_data.csv")
df.drop("meal_name",axis=1,inplace=True)
df["meal_name"]=df["cooking_method"]+ " " +df["meal_type"]+" "+df["diet_type"]
df["goal_label"] = df["cal_balance"].apply(
    lambda x: "weight loss" if x < 0 else "weight gain"
)
```

```python
col=["Age","Gender","Weight (kg)","Height (m)","BMI","Daily meals frequency","meal_name","meal_type","diet_type","cooking_method","Carbs","Proteins","Fats"
,"Calories","Workout_Type","Calories_Burned","cal_balance","goal_label"]
df_meal=df[col].copy()
```

```python
df_meal.info()
```

```python
df["cooking_method"].value_counts()
```

```python
df_meal["meal_name_raw"] = df_meal["meal_name"]
df_meal["meal_type_raw"] = df_meal["meal_type"]
df_meal["diet_type_raw"] = df_meal["diet_type"]
df_meal["cooking_method_raw"] = df_meal["cooking_method"]


```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
columns=['Gender','meal_name','meal_type','diet_type','cooking_method','Workout_Type','goal_label']
for col in columns:
    df_meal[col]=le.fit_transform(df_meal[col])
```

```python
df_meal.info()
```

```python
df_meal.to_csv("Dl_Meal.csv",index=False)
```
