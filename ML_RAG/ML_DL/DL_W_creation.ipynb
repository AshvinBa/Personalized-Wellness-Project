```python
def get_feature(exercise,feature):
    return exercise_mapping.get(exercise,{}).get(feature,None)
```

```python
df_workout["target_muscles"]=df_workout["Name of Exercise"].apply(lambda x:get_feature(x,"target_muscles"))
df_workout["focus_area"] = df_workout["Name of Exercise"].apply(lambda x:get_feature(x,"focus_area"))
df_workout["workout_type"] = df_workout["Name of Exercise"].apply(lambda x:get_feature(x,"workout_type"))
```

```python
workout_labels = ["strength", "cardio", "hiit", "yoga"]
```

```python
def encode_workout_type(workout_list):
    vec = np.zeros(len(workout_labels), dtype=int)
    for w in workout_list:
        if w in workout_labels:
            vec[workout_labels.index(w)] = 1
    return vec
```

```python
encoded_data=[]
for idx,row in df_workout.iterrows():
    name=row["Name of Exercise"]
    w_types=row["workout_type"]
    encoded_data.append({
        "Name of Exercise":name,
        "workout_type_vector":encode_workout_type(w_types)

    }
    )

```

```python
df1=pd.DataFrame(encoded_data)
workout_df = pd.DataFrame(
    df1["workout_type_vector"].to_list(),
    columns=[f"{l}" for l in workout_labels]
)
workout_df
df_workout=pd.concat([df_workout,workout_df],axis=1)
```

```python
# 0 -> Underweight
# 1 -> Normal
# 2 -> Overweight
# 3 -> Obese

def categorize_bmi_age(row):
    # For those who are childrens / Teenager.
    age = row['Age']
    bmi = row['BMI']
    if age < 19:
        if bmi < 14:
            return 0
        elif bmi < 20:
            return 1
        elif bmi < 25:
            return 2
        else:
            return 3

    # For those who are Adults
    elif age <= 65:
        if bmi < 18.5:
            return 0
        elif bmi < 25:
            return 1
        elif bmi < 30:
            return 2
        else:
            return 3

    # Elderly (65+)
    else:
        if bmi < 22:
            return 0
        elif bmi < 28:
            return 1
        elif bmi < 32:
            return 2
        else:
            return 3
     
```

```python
df_workout['BMI_Category'] = df_workout[['Age', 'BMI']].apply(categorize_bmi_age, axis=1)
```

# assigns score to each exercise based on its use in weight gain and weight loss

```python
df_workout["calories_norm"] = (
    df_workout["Burns Calories (per 30 min)"] - df_workout["Burns Calories (per 30 min)"].min()
) / (
    df_workout["Burns Calories (per 30 min)"].max() - df_workout["Burns Calories (per 30 min)"].min()
)
```

```python

df_workout["difficulty_norm"] = df_workout["Difficulty Level"] / 2

```

```python
df_workout["weight_gain_score"] = (
    0.35 * df_workout["strength"] +
    0.25 * df_workout["hiit"] +
    0.20 * df_workout["cardio"] +
    0.05 * df_workout["yoga"] +
    0.15 * df_workout["calories_norm"]
)

df_workout["weight_loss_score"] = (
    0.35 * df_workout["hiit"] +
    0.25 * df_workout["cardio"] +
    0.20 * df_workout["strength"] +
    0.05 * df_workout["yoga"] +
    0.15 * df_workout["difficulty_norm"]
)
```
